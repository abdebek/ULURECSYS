```python
def titles_references_to_csv(path: str):
    sentences = []
    with open(path, "r") as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            if line:
                l = json.loads(line)
                for aa in l['authors']:
                    if 'references' in l:
                        for rfs in l['references']:
                            sentences.append(re.sub("[\n\r]", "", l["id"] + "\t" + aa["id"] + "\t" + rfs + "\t" + re.sub("[\t]","", l["title"])) + "\n")
        lines.clear()
        sentences[-1] = re.sub("[\n]", "", sentences[-1])
    return sentences
```

```python
paths: List[str] = []
DatasetDir = "D:/Data/aminer-citation-network-ds7-v11/dblp_papers_v11"
outputDir = "D:/Data/aminer-citation-network-v11_cf/"
for dirname, _, filenames in os.walk(DatasetDir):
    for filename in filenames:
        path = os.path.join(dirname, filename)
        print("currently processing: ", path)
        sentences = titles_references_to_csv(path)
        with open(outputDir + filename + ".tsv", "w") as f:
            f.write('{}\t{}\t{}\t{}\n'.format('id','author_id', 'ref_id', 'title'))
            f.writelines(sentences)
        sentences.clear()
    #     break
    # break
                
```
